```python
from numpy import loadtxt, unique

data = loadtxt('oil-spill.csv', delimiter=',')

for i in range(data.shape[1]):
    num = len(unique(data[:, i]))
    percentage = float(num) / data.shape[0] * 100
    print("%d %d %.1f%%" % (i, num, percentage))
```

```python
from numpy import loadtxt, unique 

data = loadtxt('oil-spill.csv', delimiter=',')


for i in range(data.shape[1]):
    num = len(unique(data[:, i]))
    percentage = float(num) / data.shape[0] * 100
    print("%d %d %.1f%%" % (i, num, percentage))
    if percentage < 1.0:
        print("Column %d has only %d unique values, which is %.1f%% of the total rows." % (i, num, percentage))
```

```python
from pandas import read_csv

df = read_csv('oil-spill.csv', header=None)
print(df.shape)

counts = df.nunique()

to_del = [i for i, v in enumerate(counts) if (float(v) / df.shape[0] * 100) < 1.0]
print(to_del)

df.drop(to_del, axis=1, inplace=True)
print(df.shape)
```
